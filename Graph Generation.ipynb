```python
all_join = grph.join_all_data(filter_users=True, refresh=True)
```

```python
all_join.info()
```

```python
runner_count = all_join.groupby(["runner_name", 'is_il'])['id_runs'].count().unstack('is_il').fillna(0)
runner_count.rename({True: 'IL', False: 'Full Game'}, axis=1, inplace=True)
runner_count.loc[:, ('total_count')] = runner_count['IL'] + runner_count['Full Game']
runner_count = runner_count.sort_values('total_count', ascending=False)

runner_count.head()
```

```python
rc = runner_count.sort_values(ascending=True).plot.barh(title="Top Run Submitters")
curr_date = datetime.utcnow().strftime('%Y-%m-%d')
rc.bar_label(rc.containers[0])
rc.annotate(f"Generated on {curr_date}", xy=(1.0,-0.1), xycoords="axes fraction", ha="right", va="center", fontsize=8)
plt.savefig(grph.CHART_PATH / "Top_Submitter_Test.png", format="png", bbox_inches="tight")
```

```python
rc = runner_count['Full Game'].sort_values(ascending=False)[:10].sort_values(ascending=True).plot.barh(title="Top Fullgame Submitters")
curr_date = datetime.utcnow().strftime('%Y-%m-%d')
rc.bar_label(rc.containers[0])
rc.annotate(f"Generated on {curr_date}", xy=(1.0,-0.1), xycoords="axes fraction", ha="right", va="center", fontsize=8)
plt.savefig(grph.CHART_PATH / "Top_Fullgame_Submitter_Test.png", format="png", bbox_inches="tight")
```

```python
rc = runner_count['IL'].sort_values(ascending=False)[:10].sort_values(ascending=True).plot.barh(title="Top IL Submitters")
curr_date = datetime.utcnow().strftime('%Y-%m-%d')
rc.bar_label(rc.containers[0])
rc.annotate(f"Generated on {curr_date}", xy=(1.0,-0.1), xycoords="axes fraction", ha="right", va="center", fontsize=8)
plt.savefig(grph.CHART_PATH / "Top_IL_Submitter_Test.png", format="png", bbox_inches="tight")
```

```python
# Top 5 levels per category
runs = grph.join_all_data(filter_users=True, refresh=True)
il_counts = runs.groupby(["short_name", "Categories"])["id_runs"].count().unstack("Categories").fillna(0)
il_counts.info()
```

```python
top_any = il_counts['Any%'].sort_values(ascending=False)[:10].sort_values(ascending=True).plot.barh(title="Top Any% ILs")
curr_date = datetime.utcnow().strftime('%Y-%m-%d')
top_any.bar_label(top_any.containers[0])
top_any.annotate(f"Generated on {curr_date}", xy=(1.0,-0.1), xycoords="axes fraction", ha="right", va="center", fontsize=8)
plt.savefig(grph.CHART_PATH / "Top_IL_Any%_Test.png", format="png", bbox_inches="tight")
```

```python
top_at = il_counts['All Toppins'].sort_values(ascending=False)[:10].sort_values(ascending=True).plot.barh(title="Top All Toppins ILs", color="C1")
curr_date = datetime.utcnow().strftime('%Y-%m-%d')
top_at.bar_label(top_at.containers[0])
top_at.annotate(f"Generated on {curr_date}", xy=(1.0,-0.1), xycoords="axes fraction", ha="right", va="center", fontsize=8)
plt.savefig(grph.CHART_PATH / "Top_IL_All_Toppins_Test.png", format="png", bbox_inches="tight")
```

```python
top_100 = il_counts['100%'].sort_values(ascending=False)[:10].sort_values(ascending=True).plot.barh(title="Top 100% ILs", color="C2")
curr_date = datetime.utcnow().strftime('%Y-%m-%d')
top_100.bar_label(top_100.containers[0])
top_100.annotate(f"Generated on {curr_date}", xy=(1.0,-0.1), xycoords="axes fraction", ha="right", va="center", fontsize=8)
plt.savefig(grph.CHART_PATH / "Top_100%_Toppins_Test.png", format="png", bbox_inches="tight")
```
